### Introduction

The dataset chosen contains salary information for Data Science roles around the world for the years of 2020 and 2021; data for the year of 2021 is mostly estimated.  While the salaries are provided in the specific currency of the location that the role is set in, there is also a column that contains the salary in USD as well.  Moreover, the experience level is listed in addition to the title that the user held when the data were collected.  A link to the chosen dataset is provided below for clarity.
https://www.kaggle.com/datasets/saurabhshahane/data-science-jobs-salaries

### Question

What is the expected amount of salary to expect for a Data Science role for entry level, and what Data Science role appears to be the most fortuitous?

### Graphs

```python
# Import the required libraries.
import warnings
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

# Silence the warnings.
warnings.simplefilter(action='ignore', category=FutureWarning)
pd.options.mode.chained_assignment = None  # default='warn'

# Load the data into an object.
df_salaries = pd.read_csv('Data Science Jobs Salaries.csv')

# Clean the data by replacing the value for the year and updating the values from series to ints.
for i in range(len(df_salaries)):
    if((df_salaries['work_year'][i]) == "2021e"):
        df_salaries['work_year'][i] = 2021
    df_salaries['salary'][i] = int(df_salaries['salary'][i])
    df_salaries['work_year'][i] = int(df_salaries['work_year'][i])

# List of highest salaries by job_title.
df_salaries[['job_title', 'salary_in_usd']].sort_values(by = 'salary_in_usd', axis = 0, ascending = False).head(10)

```

```python
# Top 10 positions ranked by salary.
df_salaries[['job_title', 'salary_in_usd']].sort_values(by = 'salary_in_usd', axis = 0, ascending = False).head(10) \
    .plot(x = "job_title", y = "salary_in_usd", kind = "bar").set_title('Highest Salary By Job Title')
```

```python
# Perform the same process but check by year.
df_salaries_2020 = df_salaries[df_salaries['work_year'] == 2020]

# Checking for the year of 2020 first.
df_salaries_2020[['job_title', 'salary_in_usd']].groupby("job_title").median() \
    .sort_values(by = 'salary_in_usd', axis = 0, ascending = False).head(10) \
    .plot(legend = False, x = "job_title", y = "salary_in_usd", kind = "pie").set_title("Highest Salary By Job Title (2020)")
```

```python
# Perform the same process but check by year.
df_salaries_2020 = df_salaries[df_salaries['work_year'] == 2021]

# Checking for the year of 2021.
df_salaries_2020[['job_title', 'salary_in_usd']].groupby("job_title").median() \
    .sort_values(by = 'salary_in_usd', axis = 0, ascending = False).head(10) \
    .plot(legend = False, x = "job_title", y = "salary_in_usd", kind = "pie").set_title("Highest Salary By Job Title (2021)")
```

```python
# What is the difference in median salary between each year?
df_salaries[['work_year','job_title', 'salary_in_usd']].groupby("work_year").median() \
    .sort_values(by = "salary_in_usd", axis = 0, ascending = False)
```

```python
# Create another graph with only entry level users.
df_salaries_entryLevel = df_salaries[df_salaries['experience_level'] == "EN"]

# Checking for the year of 2021.
df_salaries_entryLevel[['job_title', 'salary_in_usd']].sort_values(by = 'salary_in_usd', axis = 0, ascending = False).head(10) \
    .plot(legend = False, x = "job_title", y = "salary_in_usd", kind = "bar").set_title("Highest Salary For Entry Level")
```
