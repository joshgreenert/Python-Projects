```python
# Import the required libraries.
import numpy as np
import pandas as pd
```

```python
# Read the data into a pandas dataframe.
df_games = pd.read_csv('Video_Games_Sales.csv')

# Display the first ten rows of data.
df_games.head(10)
```

```python
# Find the dimensions (rows and columns)
df_games.shape
```

## The rows and columns displayed above tell us the limits of our dataset. 

```python
# Find the top five games by critic score.
df_games[['Name', 'Critic_Score']].sort_values(by = 'Critic_Score', axis = 0, ascending = False).head(5)
```

```python
# Find the number of video games in the dataframe by genre
df_games[['Name', 'Genre']].groupby('Genre').count().sort_values(by = 'Genre', axis = 0, ascending = False)
```

```python
# Find the first five games on the SNES platform.
snes_games = df_games['Platform' = 'SNES']
snes_games
```
