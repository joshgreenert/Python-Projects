```python
# Import the required libraries.
import numpy as np
import pandas as pd
```

```python
# Read the data into a pandas dataframe.
df_games = pd.read_csv('Video_Games_Sales.csv')

# Display the first ten rows of data.
df_games.head(10)
```

```python
# Find the dimensions (rows and columns)
df_games.shape
```

## The rows and columns displayed above tell us the limits of our dataset. 

```python
# Find the top five games by critic score.
df_games[['Name', 'Critic_Score']].sort_values(by = 'Critic_Score', axis = 0, ascending = False).head(5)
```

```python
# Find the number of video games in the dataframe by genre
df_games[['Name', 'Genre']].groupby('Genre').count().sort_values(by = 'Genre', axis = 0, ascending = False)
```

```python
# Find the first five games on the SNES platform.
df_games[df_games['Platform'] =='SNES'].sort_values('Year_of_Release', ascending = True).head(5)
#df_games[['Name', 'Platform', 'Year_of_Release']].sort_values(['Platform', 'Year_of_Release'], ascending=[True, False]).head(5)
```

```python
# Find the five publishers with the highest total sales (calculate the total global sales for each publisher).
df_games[['Publisher', 'Global_Sales']].groupby('Publisher').sum().sort_values('Global_Sales', ascending = False).head(5)
```

```python
# Create a new column in the data frame that calculates the percentage of global sales from North America.
# Display the first five rows of the new data frame.
NA_Percent_Column = []
for i in range(len(df_games)):
    NA_Percent = round((df_games['NA_Sales'][i] / df_games['Global_Sales'][i]) * 100, 2)
    NA_Percent_Column.append(NA_Percent)

df_games['NA_Percent'] = NA_Percent_Column
df_games.head(5)
```

```python
# Find the number NaN entries (missing data values) in each column.
df_games[df_games.columns[df_games.isnull().any()]].isnull().sum()
```

```python
# Calculate the median score of all video games.  Some user score entries are non-numerical that cannot be converted to float. 
# Find and replace this string with NaN and then colculate the median.  Then, replace all NaN entries with the median value.
Median_Score = 0
median_count = 0

# Loop the range of values to get the median score.
for i in range(len(df_games)):
    # check if num == num for NaN and if tbd is listed.
    if(df_games['User_Score'][i] != df_games['User_Score'][i] or df_games['User_Score'][i] == 'tbd'):
        continue
    else:
        Median_Score = float(df_games['User_Score'][i]) + Median_Score
        median_count += 1
Median_Score = round(Median_Score / median_count, 2)

# Use the same loop as before to update all non-numeric items to the median score.
for i in range(len(df_games)):
    # check if num == num for NaN and if tbd is listed.
    if(df_games['User_Score'][i] != df_games['User_Score'][i]):
        df_games['User_Score'] = Median_Score
        
df_games.head(10)
```
