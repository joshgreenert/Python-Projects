## Import Data

```python
# Import data as a pandas data.frame.
df_movies = pd.read_csv('imdb_top_1000.csv')

# Display the first three rows of data.
df_movies.head(3)
```

## Explore the data

```python
# Check for missing data.
df_movies.isna().sum()
```

```python
# Get the top five movies by meta score.
df_movies.sort_values(by='Meta_score', axis = 0, ascending=False).head(5)
```

```python
# Get top five movie titles by metascore
df_movies[['Series_Title', 'Meta_score']].sort_values(by='Meta_score', axis=0, ascending=False).head(5)
```

```python
# Find the top five directors by mean IMDB rating.
df_movies[['Director', 'IMDB_Rating']].groupby('Director').mean() \
    .sort_values(by='IMDB_Rating', axis = 0, ascending=False).head(5)
```
