### Prepare the data

```python
# Import the required libaries for plotting.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Pull the data into a dataframe object.
df_hotdog = pd.read_excel("hotdog-contest-winners.xlsm")

# Update the country values to be lesser digits.
for index, row in df_hotdog.iterrows():
    if row['Country'] == "United States":
        df_hotdog.at[index,'Country'] = "US"
    elif row['Country'] == "Mexico":
        df_hotdog.at[index,'Country'] = "MX"
    elif row['Country'] == "Germany":
        df_hotdog.at[index,'Country'] = "DE"
    elif row['Country'] == "Japan":
        df_hotdog.at[index,'Country'] = "JP"

df_hotdog.head(5)
```

### Python - Bar Chart


```python
# Set a portion of the data to plot.
df_hotdog_country = df_hotdog[['Country','Dogs eaten']]
ax = df_hotdog_country.plot.bar(y='Dogs eaten', x='Country',rot=0, figsize=(20,10))
```

### Python - Stacked Bar Chart


```python
ax = df_hotdog.plot.bar(rot=0, figsize=(20,10), stacked=True)
```

### Python - Pie Chart

```python
# Group the values by country to plot the pie chart.
df_grouped_country = df_hotdog_country.groupby(['Country']).sum()
ax = df_grouped_country.plot.pie(y='Dogs eaten', x='Country',rot=0)
```

### Python - Donut Chart

```python
# Create a pie chart.
plt.pie(df_grouped_country['Dogs eaten'], autopct='%1.1f%%', pctdistance=0.85, explode = (0.05, 0.05, 0.05, 0.05))
  
# draw circle.
centre_circle = plt.Circle((0, 0), 0.70, fc='white')

fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Display donut chart.
plt.title('Winner by Country')
plt.show()
```
